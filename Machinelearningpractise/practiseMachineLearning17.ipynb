```python
df_obj=df.select_dtypes([object,bool])
df_obj
```

```python
df_num=df.select_dtypes(['int64','float64'])
df_num
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_obj.columns

for i in df_obj:
    df_obj[i]=le.fit_transform(df_obj[i])
```

```python
df_obj
```

```python
df_sum=pd.concat([df_num,df_obj],axis=1)
df_sum
```

```python
X=df_sum.drop('survived',axis=1)
Y=df_sum['survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
dict={"Actual op":Y_test,'Predict op':y_pred}
df2=pd.DataFrame(dict)
df2
```

```python
lr.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
TP,FP,FN,TN=confusion_matrix(Y_test,y_pred).flatten()
```

```python
print("True positive",TP)
print("False positve",FP)
print("False negative",FN)
print('True negative',TN)
```
