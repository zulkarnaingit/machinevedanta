```python
df['rate']=df['rate'].str.replace(r'5', ' ') #replace 5 by empty space
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
# Create a bar plot to visualize 'rate' for each 'name'
sb.countplot(data=df,y='rate')

plt.title('number of rating given by customers ',color='green',fontsize=40,fontname='Arial')

plt.xlabel('number of rating')
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(30,30))
plt.title('average percentage rating in given by customers ',color='Red',fontsize=40,fontname='Arial')
df['rate'].value_counts().plot(kind='pie',autopct='%.2f%%')
plt.legend()
```
