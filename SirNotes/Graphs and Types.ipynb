# SCATTER PLOT

```python
#basic Library function
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
import os
os.getcwd()
```

```python
#to load dataset .csv
df=pd.read_csv("homeprices.csv")
df
```

```python
#to change the size of the page and format of the page we can use inbuilt method figure()
plt.figure(figsize=(4,5),linewidth=7,facecolor='g',edgecolor='b')#parameters used are figsize(length,width),linewidth,facecolor,edgecolor
#scatterplot: input and output
sb.scatterplot(data=df,x='Area',y='Price',color='red',s=150,marker='^')
#s stands for size of datapoint and the values is in pixels
#marker:- It is used to changes the shape of datapoint
        # In marker 's' stands for square
plt.xticks(df['Area'])#xticks it shows us the exact numbers on the graph for X-axis
plt.yticks(df['Price'])#yticks it shows us the exact numbers on the graph for Y-axis
plt.title("Area vs Price",color='blue', fontsize=20,fontname="Arial")
#to define the title of X-axis
plt.xlabel("Area of House in Square.ft", fontsize=20)
#to define the title of Y-axis
plt.ylabel("Price of House in Lacs",fontsize=18)
#to save a graph in image format(.png), we use in-built method savefig()
#This savefig() is define in matplotlib.pyplot class
plt.savefig("graph1.png") #permanent new file name and save in by default path.
plt.show()
```

# LINE GRAPH

```python
sb.lineplot(data=df,x='Area',y='Price',color='red',linestyle='dotted',linewidth=7)
#linestyle is use to change the style of the line such as dotted,dash
#linewidth is use to size of the line
plt.xticks(df['Area'])#xticks it shows us the exact numbers on the graph for X-axis
plt.yticks(df['Price'])#yticks it shows us the exact numbers on the graph for Y-axis
plt.title("Area vs Price",color='green', fontsize=40,fontname='Arial')#it shows the title() of the graph
#To define the label on X-axis
plt.xlabel("Area of House in square.ft",fontsize=20)
#To define the label on Y-axis
plt.ylabel("Price of houses in Lacs",fontsize=20)
#to save the file in .format 
#plt.savefig("graph2.pdf")
plt.show()
```

```python
#using line plot and scatterplot at the same time
sb.lineplot(data=df,x='Area',y='Price',color='red')
sb.scatterplot(data=df,x='Area',y='Price',s=140)
plt.xticks(df['Area'])
plt.yticks(df['Price'])
plt.title("Area vs Price", fontsize=20, fontname='Arial')
plt.xlabel("Area in square feet")
plt.ylabel("Prices of house in lacs")
#plt.savefig("graph3.pdf")
plt.show()
```

# 3.HISTOGRAM

```python
#Histogram:- It is also known as frequency graph
#It is used always in contineous numeric datatype.
#contineous numerical data means which we cannot count but we can measure the data
#In histogram we can take input on X-axis or Y-axis.
```

```python
dict={'Age':[5,6,6,6,6,7,7,7,8,8,8,8,8,9,9,10,10,10,10,10,10]}
#convert dict into dataframe
df1=pd.DataFrame(dict)
df1
#how to count values 
df1['Age'].value_counts()
```

```python
sb.histplot(data=df1,x='Age')
plt.title("Age Count of the People", fontsize=15)
plt.xlabel("Age Group")
plt.ylabel("Count of People")
plt.show()
```

```python
#if you want to show the data individually with the help of histogram
plt.hist(df1['Age'],rwidth=0.7)
plt.show()
```

# BOXPLOT

# BAR GRAPH

```python
#taking a  new dataset name titanic
df6=pd.read_csv('titanic.csv')
df6.head()
```

```python
#Bar means vertical bar in which on X-axis: input and on Y-axis: frequency
#how many passangers are travelling 1st class, 2nd class, 3rd class
plt.figure(figsize=(4,3))
df6['Pclass'].value_counts().plot(kind='bar',color=['red','green','yellow'])#way of giving multiple colors in BarGraph
f=df6['Pclass'].value_counts()
plt.yticks(f)
plt.xlabel("Passanger Class")
plt.ylabel("No. of Passanger")
plt.show()
```

```python
#Barh means Horizontal bar in which X-axis: frequency and on Y-axis: input
plt.figure(figsize=(6,3))
df6['Pclass'].value_counts().plot(kind='barh',color=['red','green','yellow'])#way of giving multiple colors in BarGraph
f=df6['Pclass'].value_counts()
plt.xticks(f)
plt.xlabel("No. of Passanger")
plt.ylabel("Passenger Class")
plt.show()
```

# PIE GRAPH

```python
myexplode=[0.2,0,0]
df6['Pclass'].value_counts().plot(kind='pie',autopct='%.2f%%',colors=['yellow','green','red'], explode=myexplode)
plt.legend()
plt.show()
```

# COUNT PLOT

```python
sb.countplot(data=df6,x='Pclass')
f=df6['Pclass'].value_counts()
plt.yticks(f)
plt.show()
```

```python
#how to calculate form class with secondary information
#how many males travel in Pclass 1,2,3 and females in Pclass 1,2,3
sb.countplot(data=df6,x='Pclass',hue='Sex') #hue is the keyword for secondary
plt.show()
```
