**Sample DataSet**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
columns = ['sepal_length','sepal_width','petal_length','petal_width','species']
# df.rename(columns=['sepal_length','sepal_width','petal_length','petal_width','species'],axis=1,inplace=True)
```

```python
df.columns = columns
```

```python
df
```

```python
df.columns
```

```python
features = df[['sepal_length','sepal_width','petal_length','petal_width']]
target = df['species']
```

```python
features
```

```python
target.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbl_encode = LabelEncoder()
```

```python
encoded_labels = lbl_encode.fit_transform(target)
```

```python
encoded_labels
```

```python
features
```

```python
encoded_labels
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
scaled_features = sc.fit_transform(features)
```

```python
scaled_features
```

```python
df.describe()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca_features = pca.fit_transform(scaled_features)
```

```python
pca_features
```

```python
import matplotlib.pyplot as plt
```

```python
features
```

```python
plt.scatter(features['sepal_length'],features['sepal_width'],c=encoded_labels)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('ORiginal Data')
plt.show()
```

```python
plt.scatter(pca_features[:,0],pca_features[:,1],c=encoded_labels)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Data After PCA')
plt.show()
```
