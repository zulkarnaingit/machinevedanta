```python
#handling object type of columns
#taking make column in 1 list

m = df['make'].unique()
print(m)
```

```python
df.head(10)
```

```python
# now splitting the dataframe into 2 dataframe
# 1.Dataframe will hold all numeric type of columns with data
# 2.Dataframe will hold all object type of columns with data
# we use inbuilt method select_dtypes()

# create a new dataset df_num dataframe which will hold all numeric type of data
df_num = df.select_dtypes(['float64','int64'])
df_num.dtypes
```

```python
df_num.head(10)
```

```python
# create a new dataset df_obj dataframe which will hold all object type of data
df_obj = df.select_dtypes(['object'])
df_obj.dtypes
```

```python
df_obj.head(10)
```

```python
#applying labelencoder to convert all object type data into number
#calling inbuilt class labelencoder

from sklearn.preprocessing import LabelEncoder

#create an object for labelEncoder

le = LabelEncoder()
df_obj.columns
for col in df_obj:
    df_obj[col] = le.fit_transform(df_obj[col])
```

```python
df_obj.head(10)
```

```python
m1 = df_obj['make'].unique()
m1
```

```python
#making a dictionary and convertingit into dataframe

dict = {'cars_name':m,'cars_numbers':m1}
df1 = pd.DataFrame(dict)
df1
```

```python
#merging both the dataframe (df_num,df_obj) into 1 main dataframe
#use inbuilt pandas method concat()
#syntax:- pd.concat([dataframe1,dataframe2],axis=1) #axis=1 for dataframe to start from left to right

df_new = pd.concat([df_num,df_obj],axis=1)
```

```python
df_new.head(10)
```

```python
#selecting input and output

X = df_new[['symboling','normalized-losses','width','height','engine-size','horsepower','city-mpg','highway-mpg','make','fuel-type','body-style','drive-wheels','engine-location','engine-type']]  

#X = df_new.drop('price',axis=1)

Y = df_new["price"]
```

```python
#train test split

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

#create an object for the standard scaler

ss=StandardScaler()

```

```python
#appling scailing on X_train we use fit_transform()

X_train = ss.fit_transform(X_train)


#appling scaling on X_test,we use .transform()

X_test = ss.transform(X_test)
```

```python
#performing the model with linear regression
# we have to import linear regression from linear_model

from sklearn.linear_model import LinearRegression

#create an object for LinearRegresssion

lr = LinearRegression()
```

```python
#lets train the model with 70% of data
# we use .fit() method of linear regression class

lr.fit(X_train,Y_train)
```

```python
#checking the slope of the model

lr.coef_
print('Slope of input: ',lr.coef_)
```

```python
# lets just find the intercept/constant

lr.intercept_
print('intercept of input: ',lr.intercept_)
```

```python


y_pred = lr.predict(X_test)
print('Predicted output: ',y_pred)
```

```python
dict = {'Actual Output':Y_test,'Predicted output':y_pred}

#convert dict into dataframe


df10 = pd.DataFrame(dict)

df10
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Mean Squared Error: ',mean_squared_error(Y_test,y_pred))
```

```python
mse = mean_squared_error(Y_test,y_pred)
```

```python
rmse = np.sqrt(mse)
print('Root Mean Squared Error: ',rmse)
```

```python
#checking for the score of model: r2_score
# use inbuilt function r2_score() which is defined in r2_score

from sklearn.metrics import r2_score

print('r2_score: ',r2_score(Y_test,y_pred))
```

```python
residual_error = Y_test-y_pred
print('Residual Error for each datapoints: ',residual_error)
```

```python
plt.scatter(y_pred,residual_error)
#plt.plot(y_pred,residual_error,color='red')
plt.xlabel('Predicted')
plt.ylabel('Residual Error')
plt.show()
```
